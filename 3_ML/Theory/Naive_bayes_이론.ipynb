### - Naive bayes classifier 배경

##### 입력변수 : 겉보기 날씨와 습도
##### 출력변수 : 플레이 테니스 

 ex.) 겉보기 날씨와 습도가 고정되어 있다고 할 떄, 가질 수 있는 패턴은 두 가지

    -> 테니스를 친다., 테니스를 치지 않는다.


- 설명변수와 종속변수는 분리해서 생각.
